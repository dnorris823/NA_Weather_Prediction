```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('full_dataframe.csv', compression='zip')
```

```python
df_seattle = df[df['city_Seattle'] == 1]
```

```python
sns.set(color_codes=True)
sns.regplot(x=df_seattle['humidity'], y=df_seattle['temperature'])
```

```python
sns.set(color_codes=True)
sns.regplot(x=df_seattle['pressure'], y=df_seattle['temperature'])
```

```python
sns.set(color_codes=True)
sns.regplot(x=df_seattle['wind direction'], y=df_seattle['temperature'])
```

```python
sns.set(color_codes=True)
sns.regplot(x=df_seattle['wind speed'], y=df_seattle['temperature'])
```
