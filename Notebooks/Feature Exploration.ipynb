```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('full_dataframe.csv', compression='zip')
```

The next few cells are find the correlation and relationships between the temperature and the other continous variables for a single city.

The next section of cells will be about exploring the data when it is downsampled from hourly recordings to monthly recordings

```python
df_sea_temp_mon = pd.DataFrame()
df_sea_temp_mon['datetime'] = pd.to_datetime(df_seattle['datetime'])
df_sea_temp_mon['temperature'] = df_seattle['temperature']
df_sea_temp_mon = df_sea_temp_mon.set_index('datetime')
df_sea_temp_mon['temperature'] = df_sea_temp_mon['temperature'].asfreq('M')
df_sea_temp_mon = df_sea_temp_mon.dropna()
```

```python
df_sea_hum_mon = pd.DataFrame()
df_sea_hum_mon['datetime'] = pd.to_datetime(df_seattle['datetime'])
df_sea_hum_mon['humidity'] = df_seattle['humidity']
df_sea_hum_mon = df_sea_hum_mon.set_index('datetime')
df_sea_hum_mon['humidity'] = df_sea_hum_mon['humidity'].asfreq('M')
df_sea_hum_mon = df_sea_hum_mon.dropna()
```

```python
df_sea_press_mon = pd.DataFrame()
df_sea_press_mon['datetime'] = pd.to_datetime(df_seattle['datetime'])
df_sea_press_mon['pressure'] = df_seattle['pressure']
df_sea_press_mon = df_sea_press_mon.set_index('datetime')
df_sea_press_mon['pressure'] = df_sea_press_mon['pressure'].asfreq('M')
df_sea_press_mon = df_sea_press_mon.dropna()
```

```python
df_sea_wdir_mon = pd.DataFrame()
df_sea_wdir_mon['datetime'] = pd.to_datetime(df_seattle['datetime'])
df_sea_wdir_mon['wind direction'] = df_seattle['wind direction']
df_sea_wdir_mon = df_sea_wdir_mon.set_index('datetime')
df_sea_wdir_mon['wind direction'] = df_sea_wdir_mon['wind direction'].asfreq('M')
df_sea_wdir_mon = df_sea_wdir_mon.dropna()
```

```python
df_sea_wspeed_mon = pd.DataFrame()
df_sea_wspeed_mon['datetime'] = pd.to_datetime(df_seattle['datetime'])
df_sea_wspeed_mon['wind speed'] = df_seattle['wind speed']
df_sea_wspeed_mon = df_sea_wspeed_mon.set_index('datetime')
df_sea_wspeed_mon['wind speed'] = df_sea_wspeed_mon['wind speed'].asfreq('M')
df_sea_wspeed_mon = df_sea_wspeed_mon.dropna()
```

```python
df_sea_mon = pd.DataFrame()
df_sea_mon['datetime'] = pd.to_datetime(df_seattle['datetime'])
df_sea_mon['temperature'] = df_seattle['temperature']
df_sea_mon = df_sea_mon.set_index('datetime')
df_sea_mon['temperature'] = df_sea_mon['temperature'].asfreq('M')
df_sea_mon = df_sea_mon.dropna()
```

```python
df_sea_mon['humidity'] = df_sea_hum_mon['humidity']
df_sea_mon['pressure'] = df_sea_press_mon['pressure']
df_sea_mon['wind speed'] = df_sea_wspeed_mon['wind speed']
df_sea_mon['wind direction'] = df_sea_wdir_mon['wind direction']
```
